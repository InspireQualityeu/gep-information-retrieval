```python
import pandas as pd
import os
import re


# Specify the directory path. Becarefull with slash and backslash.

directory_path = r'define_directory_for_reading'

# Check if the directory exists
if os.path.exists(directory_path):
    # List all files in the directory with .txt extension
    txt_files = [f for f in os.listdir(directory_path) if f.endswith('.txt')]
    
    # Initialize lists to store data
    text_data = []
    pdf_names = []
    
    # Loop through each .txt file and read its contents
    for txt_file in txt_files:
        file_path = os.path.join(directory_path, txt_file)
        with open(file_path, 'r', encoding='utf-8') as file:
            file_contents = file.read()
            # Append the file name (without extension) to pdf_names
            pdf_name_without_extension = os.path.splitext(txt_file)[0]
            pdf_names.append(pdf_name_without_extension)
            # Append the file contents to text_data
            text_data.append(file_contents)

    # Create a DataFrame
    df = pd.DataFrame({'PDF_Name': pdf_names, 'Text': text_data})
    
    # Print the DataFrame
    print(df)
else:
    print(f"The directory {directory_path} does not exist.")

```

```python
###The time plan of the current plan - Indicator 5

# Define a regular expression pattern to match dates in the format "xxxx-xxxx"

date_pattern_v1 = r'\s*?\d{4}\s*?[\u002D\u058A\u05BE\u2010\u2011\u2012\u2013\u2014\u2015\u2E3A\u2E3B\uFE58\uFE63\uFF0D]\s*?\d{4}\s*?'

# Function to extract date from text
def extract_date_v1(text):
    match = re.search(date_pattern_v1, text)
    if match:
        return match.group(0)
    else:
        return None

# Apply the function to create a new 'time frame' column
df['Indicator_5_the_time_frame_of_the_current_plan'] = df['Text'].apply(extract_date_v1)
```
